#Text Preprocessing

#Case Folding , stopwords , punctuation

````python
punctuation =  set(string.punctuation)
stop_words = set(stopwords.words("english"))
stop_words.update(["br", "href","amazon","product","counter"])
````

````python

def Clean_Text(review):
  x = review
  x = re.sub(r'https?:\S', '', x)
  x = re.sub(r'{link}', '', x)
  x = re.sub(r'@[\w]', '', x)
  x = re.sub(r'[^A-Za-z0-9]+', ' ', x)
  x = re.sub(r'\b[0-9]+\b', '', x)
  x = re.sub(r'&[a-z]+', '', x)
  x = ''.join(w for w in x if w not in punctuation)
  x = x.replace('[^a-zA-Z#]',' ')

  x = [w.lower() for w in x.split() if w.lower() not in stop_words]
  x = [w for w in x if len(w) > 3]
  return x
````

````python
df['Clean_Text'] = df['Review'].apply(lambda x: Clean_Text(x) if isinstance(x, str) else '')
````

# Lemmatization

````python
nltk.download('wordnet')
````

````python
from nltk.corpus import wordnet as wn
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import pandas as pd

# Initialize variables and objects
lemmatizer = WordNetLemmatizer()

# A dictionary mapping part-of-speech tags to WordNet part-of-speech constants.
tag_dict = {"J": wn.ADJ,
            "N": wn.NOUN,
            "V": wn.VERB,
            "R": wn.ADV}

# Maps a part-of-speech tag (e.g., 'NN') to its WordNet equivalent.
def mapToWordEq(tag):
    return tag_dict.get(tag[0].upper(), None)

# Function to lemmatize a word given its tuple (word, pos_tag)
def lemmatize_tupla_word_postag(tupla):
    tag = mapToWordEq(tupla[1])
    return lemmatizer.lemmatize(tupla[0], tag) if tag is not None else tupla[0]

# Function to perform bag of words processing on a sentence
def bag_of_words(sentence):
    original_words = word_tokenize(sentence)
    tagged_words = nltk.pos_tag(original_words)
    lemmatized_words = [lemmatize_tupla_word_postag(ow) for ow in tagged_words]

    return lemmatized_words
````

#Sentiment Analysis


````python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')
````

````python
analyzer = SentimentIntensityAnalyzer()

def get_sentiment(text):
    scores = analyzer.polarity_scores(text)

    if scores['compound'] >= 0.05 :
        sentiment = 2
    elif scores['compound'] <= -0.05:
        sentiment = 0
    else:
        sentiment = 1

    return sentiment


df['sentimentPolarity'] = df['Clean_Text'].apply(get_sentiment)

df
````

````python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(df['SentimentManiual'], df['sentimentPolarity']))
````

````python
from sklearn.metrics import classification_report, accuracy_score

accuracy = accuracy_score(df['SentimentManiual'], df['sentimentPolarity'])

print("Accuracy:", accuracy)

print(classification_report(df['SentimentManiual'], df['sentimentPolarity']))
````

````python
def eqNeg (cmp):
  return (cmp-(-1))*((2-1)/(-0.05+1))+1

def eqPos(cmp):
  return (cmp-(0.05))*((5-4)/(1-(0.05)))+4
````

````python
analyzer = SentimentIntensityAnalyzer()

def get_sentiment(text):
    scores = analyzer.polarity_scores(text)

    if scores['compound'] >= 0.05 :
        sentiment = round(eqPos(scores['compound']))
    elif scores['compound'] <= -0.05:
        sentiment = round(eqNeg(scores['compound']))
    else:
        sentiment = 3

    return sentiment

df['scores']= df['Clean_Text'].apply(analyzer.polarity_scores)

df['sentiment'] = df['Clean_Text'].apply(get_sentiment)

df
````

````python
df.rename(columns={'sentiment': 'sentimentScore'}, inplace=True)
df = df.drop(columns=['scores'],axis=1)
````

````python
from sklearn.metrics import classification_report, accuracy_score
accuracy = accuracy_score(df['Score'], df['sentimentScore'])
print("Accuracy:", accuracy)

print(classification_report(df['Score'], df['sentimentScore']))
````
